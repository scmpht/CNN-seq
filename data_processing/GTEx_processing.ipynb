```python
import pandas as pd
import anndata as ad
import numpy as np
```

```python
gtc = pd.read_csv("/nobackup/scmpht/CNN-seq/data/raw/GTEx_Analysis_v10_RNASeQCv2.4.2_gene_tpm.gct", skiprows=2, sep="\t", low_memory=True, index_col=0)
gene_names = gtc['Description']
gtc.drop(columns=['Description'], inplace=True)
gtc
```

```python
metadata = pd.read_csv("/nobackup/scmpht/CNN-seq/data/raw/GTEx_Analysis_v10_Annotations_SampleAttributesDS.txt", sep="\t", index_col=0, low_memory=False)
metadata = metadata[['SMTS', 'SMTSD', 'SMNABTCH', 'SMNABTCHT', 'SMNABTCHD', 'SMGEBTCH', 'SMGEBTCHD', 'SMGEBTCHT', 'SMCENTER']]
metadata = metadata.loc[gtc.columns]
metadata
```

```python
metadata['SMTS'].value_counts()
```

```python
adata = ad.AnnData(X=gtc.T, obs=metadata)
adata.var['Symbol'] = gene_names
adata.write_h5ad("/nobackup/scmpht/CNN-seq/data/processed/GTEx_raw.h5ad")
adata[::10, ::10].write_h5ad("/nobackup/scmpht/CNN-seq/data/processed/dummy_GTEx_raw.h5ad")
```

```python
# adata = ad.read_h5ad("/nobackup/scmpht/CNN-seq/data/processed/dummy_GTEx_raw.h5ad")
adata.X = np.log2(adata.X + 1) # Log transform
adata = adata[:, adata.X.var(axis=0) > 0.5] # Remove low variance genes, this will include all 0.
adata = adata[:, adata.X.mean(axis=0) > 1] # Remove genes with low mean count and suscpetible to noise

adata.write_h5ad("/nobackup/scmpht/CNN-seq/data/model_ready/GTEx.h5ad")
adata[::10, ::10].write_h5ad("/nobackup/scmpht/CNN-seq/data/model_ready/dummy_GTEx.h5ad")
adata
```
